```python
import pandas as pd            
```

```python
# read csv file
df = pd.read_csv('../Bryan/driver_info.csv')
df
```

```python
# convert Salary column to string
df['Salary'] = df['Salary'].astype(str)
df
```

```python
# drop the M from the Salary column
df['Salary'] = df['Salary'].str.replace('M', '')
df
```

```python
# convert Salary column to float
df['Salary'] = df['Salary'].astype(float)
df
```

```python
# arrange table by Salary descending
df = df.sort_values(by=['Salary'], ascending=False)
df
```

```python
# add $ to the beginning of the Salary column and M to the end
df['Salary'] = '$' + df['Salary'].astype(str) + 'M'
df
```

```python
# save to csv file
df.to_csv('../Bryan/driver_info.csv', index=False)
```
