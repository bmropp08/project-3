```python
# import dependencies
import pandas as pd
```

```python
# read in csv file
df = pd.read_csv("../Formula1/drivers.csv")
df.head()
```

```python
# drop unnecessary columns
df = df.drop(columns=["driverId", "driverRef", "number", "code", "dob", "url"])
df.head()
```

```python
# rename columns
df = df.rename(columns={"forename": "First", "surname": "Last", "nationality": "Nationality"})
df.head()
```

```python
# combine first and last name columns
df["Driver"] = df["First"] + " " + df["Last"]
df.head()
```

```python
# drop first and last name columns
df = df.drop(columns=["First", "Last"])
df.head()
```

```python
# move last column to first
cols = list(df.columns)
cols = [cols[-1]] + cols[:-1]
df = df[cols]
df.head()
```

```python
# save to csv
df.to_csv("../Formula1/drivers_clean.csv", index=False)
```
