```python
from webdriver_manager.chrome import ChromeDriverManager

from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
pip install webdriver_manager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=True)
```

```python
# Visit the Formula 1 site for salaries
url = 'https://www.sportingnews.com/us/formula-1/news/how-much-f1-drivers-paid-salaries-teams-2023/kr2jqdm7hofmbpyohlckmuy5'
browser.visit(url)
```

```python
html = soup(browser.html, 'html.parser')
```

```python
table = html.find('div', class_='table-container')
table
```

```python
# Create a dataframe from the table 
df = pd.read_html(str(table))[0]
df
```

```python
#change column names
df.columns = ['Driver', 'Team', 'Salary', 'Age']
df
```

```python
#export as csv  
df.to_csv('f1_salaries.csv', index=False)
```
