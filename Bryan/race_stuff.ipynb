```python
#import dependencies
import pandas as pd
```

```python
#Save path to data set in a variable
races = "../Jason/Formula1/races.csv"
results = "../Jason/Formula1/results.csv"
drivers = "../Jason/Formula1/drivers.csv"
```

```python
races_df = pd.read_csv(races)
races_df.head()
```

```python
#filter rows with year 2023
races_df['year'] = pd.to_datetime(races_df['date']).dt.year       
desired_year = 2023
races_2023 = races_df[races_df['year'] == desired_year]
races_2023.head()
```

```python
results_df = pd.read_csv(results)
results_df.head()
```

```python
drivers_df = pd.read_csv(drivers)
drivers_df.head()
```

```python
#merge results_df and races_2023 on raceId
races_results = pd.merge(results_df, races_2023, on='raceId')
races_results.head()
```
